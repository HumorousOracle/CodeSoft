```python
#calculating average rating of movies
avg_rating = df_prediction.groupby(['Year', 'Genre'])['Rating'].mean().reset_index()

#get top 5 genres
top_genres= df_prediction['Genre'].value_counts().head(5).index

#create the line plot with plotly
fig= px.line(avg_rating, x='Year', y='Rating', color= "Genre")

#adding title and hue
fig.update_layout(title='Average rating for top genres', xaxis_title= 'Year', yaxis_title= 'Average rating')
fig.show()
```

```python
#This histogram shows the distribution of ratings and its probable density

rating_fig = px.histogram(df_prediction, x='Rating', histnorm='probability density', nbins=40)
rating_fig.update_layout(title='Rating Distribution',  title_x=0.5, title_pad=dict(t=20), title_font=dict(size=20),  xaxis_title= 'Rating', yaxis_title= 'Probability')
rating_fig.show()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error, r2_score
```

```python
df_prediction.drop('Name', axis =1, inplace = True)
```

```python
#Grouping the columns by their average rating and then creating a new feature
genre_mean_rating = df_prediction.groupby('Genre')['Rating'].transform('mean')
df_prediction['Genre_mean_rating']=  genre_mean_rating

director_mean_rating = df_prediction.groupby('Director')['Rating'].transform('mean')
df_prediction['Director_encoded'] = director_mean_rating

actor1_mean_rating = df_prediction.groupby('Actor 1')['Rating'].transform('mean')
df_prediction['Actor1_encoded'] = actor1_mean_rating

actor2_mean_rating = df_prediction.groupby('Actor 2')['Rating'].transform('mean')
df_prediction['Actor2_encoded'] = actor2_mean_rating

actor3_mean_rating = df_prediction.groupby('Actor 3')['Rating'].transform('mean')
df_prediction['Actor3_encoded'] = actor3_mean_rating
```

```python
#predictor and target variable

x = df_prediction[['Year', 'Votes', 'Duration', 'Genre_mean_rating', 'Director_encoded', 'Actor1_encoded', 'Actor2_encoded', 'Actor3_encoded']]
y= df_prediction['Rating']
```

```python
#splitting dataset into testing and traning parts
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42)
```

**MODEL** **BUILDING**

```python
model = LinearRegression()
model.fit(x_train, y_train)
model_pred = model.predict(x_test)
```
